```python
column_names = ['creditability', 'RT0']
creditability_df = df.loc[:, column_names]
```

```python
column_names = ['consistency', 'RT0']
consistency_df = df.loc[:, column_names]
```

```python
consistency_df_1 = consistency_df[consistency_df['RT0'] == 1]
consistency_df_2 = consistency_df[consistency_df['RT0'] == 2]
consistency_df_3 = consistency_df[consistency_df['RT0'] == 3]
creditability_df_1 = creditability_df[creditability_df['RT0'] == 1]
creditability_df_2 = creditability_df[creditability_df['RT0'] == 2]
creditability_df_3 = creditability_df[creditability_df['RT0'] == 3]
```

```python
con_1 = consistency_df_1.loc[:, 'consistency']
con_2 = consistency_df_2.loc[:, 'consistency']
con_3 = consistency_df_3.loc[:, 'consistency']
cre_1 = creditability_df_1.loc[:, 'creditability']
cre_2 = creditability_df_2.loc[:, 'creditability']
cre_3 = creditability_df_3.loc[:, 'creditability']
```

# One-way Anova

```python
f_statistic, p_value = f_oneway(con_1, con_2, con_3)

print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")
```

```python
f_statistic, p_value = f_oneway(cre_1, cre_2, cre_3)

print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1)
ax.boxplot([con_1, con_2, con_3])
ax.set_xticklabels(["con_1", "con_2", "con_3"]) 
ax.set_ylabel("mean")
plt.show()
```

```python
column_names = ['consistency', 'creditability', 'RT0', 'SD21']
two_way_data = df.loc[:, column_names]
```

```python
two_way_data.groupby(["RT0", "SD21"]).count()
```

```python
data = sm.datasets.get_rdataset("Moore", "carData").data
data = data.rename(columns={"partner.status": "partner_status"})
```

```python
model = ols('conformity ~ C(fcategory) * C(partner_status)', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)
```

```python
model = ols('consistency ~ C(RT0) + C(SD21) + C(RT0):C(SD21)', data=two_way_data).fit()
	# creditability	RT0	SD21

anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)
```

```python
model = ols('creditability ~ C(RT0) + C(SD21) + C(RT0):C(SD21)', data=two_way_data).fit()
	# creditability	RT0	SD21

anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)
```
